```python
import os
import pandas as pd
import numpy as np
import sklearn
from sklearn import metrics
import matplotlib.pyplot as plt
from typing import List
from rouge import Rouge 
rouge = Rouge()

import nltk
from nltk.translate.bleu_score import corpus_bleu
from nltk.translate.bleu_score import SmoothingFunction
smooth = SmoothingFunction()
```

 # result

## classification task

```python
def print_metrics(name, y_true,y_pred, target_names:List[str]):
    report = metrics.classification_report( y_true=y_true, y_pred=y_pred, target_names=target_names, digits=4 )
    # 创建画布和子图
    fig, axs = plt.subplots(2, 1, figsize=(8, 10))
    plt.subplots_adjust(top=0.6)
    # 绘制report
    axs[0].set_title(f"GenomicLLM {name} result", fontsize=18)
    axs[0].text(0.01, 0.05, report, fontsize=12, fontfamily='monospace')
    axs[0].axis('off')
    # 绘制混淆矩阵
    cm = metrics.confusion_matrix(y_true, y_pred)
    disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
    disp.plot(ax=axs[1], cmap=plt.cm.Blues,  xticks_rotation='vertical')
    axs[1].set_title('Confusion Matrix', fontsize=16)
#     plt.show()
#     plt.savefig(f'out/metrics/result_fig_{name}.png', bbox_inches='tight', dpi=150)
    
```

```python
target_dict = {
    'sequence direction':['forward', 'reverse'], 
    'gene biotype':['other', 'protein coding gene'],
    'splice site':['donor site', 'acceptor site', 'non-splice site'],
    'regulatory':['enhancer', 'promoter', 'ocr'],
    'other':['False', 'True'],
}
# all test
task_list = [
    'sequence direction', 'gene biotype', 'proximal promoter', 'proximal tata promoter','proximal notata promoter', 
    'core promoter', 'core tata promoter',  'core notata promoter', 'transcription factor binding site', 'splice site',
    'enhancer', 'Open Reading Frame', 'regulatory', 'promoter',
    'ocr',
    ]
#GUE test
# task_list = [
#     'proximal promoter', 'proximal tata promoter','proximal notata promoter', 
#     'core promoter', 'core tata promoter',  'core notata promoter', 'transcription factor binding site', 'splice site',
#     ]
#simgle test
# task_list = ['splice site']

for task in task_list:
    if 'promoter' == task:
        df_data = df[ df["input text"].apply(lambda x: "promoter" in x  and 'core'not in x and 'proximal'not in x)] 
    else:
        df_data = df[ df["input text"].apply(lambda x: task in x.split(',')[-1]) ]
    
    print(task)
    print('dataset', len(df_data), '; ',list(df_data['generated text'].unique()))
    df_data = df_data.drop_duplicates()
    
    if task in list(target_dict.keys()):
        df_data = df_data[df_data["generated text"].isin(target_dict[task])]
    else:
        df_data["true output text"] = df_data["true output text"].astype(str)
        df_data["generated text"] = df_data["generated text"].astype(str)
        df_data = df_data[df_data["generated text"].isin(target_dict['other'])]

    print('dataset', len(df_data), '; ',list(df_data['generated text'].unique()))
    target_names = target_dict[task] if task in list(target_dict.keys()) else target_dict['other']
    df_data = df_data[df_data['true output text'].isin(target_names)]
    print(len(df_data))
    
    y_true = df_data['true output text']
    y_pred = df_data["generated text"]
    print("result:    F1_score:{:.4f}   MCC:{:.4f} ".format(metrics.f1_score(y_true=y_true, y_pred=y_pred, average='macro'),
                                                            metrics.matthews_corrcoef(y_true=y_true, y_pred=y_pred)))
    print_metrics(task, y_true, y_pred, target_names)
    
```

## generation task

```python
from nltk.translate.bleu_score import sentence_bleu

task_list = ['reading comprehension', 'annotation:description', 'complete the blank', 'protein seq', 
             'reverse sequence', 'complementary sequence', 'reverse complementary sequence']

df_result_gen = pd.DataFrame()
for task in task_list:
    print(task)
    if 'complementary sequence' == task:
        df_data = df[ df["input text"].apply(lambda x: task in x.split(",")[-1] and "reverse complementary sequence" not in x.split(",")[-1]) ]
    else:
        df_data = df[ df["input text"].apply(lambda x: task in x.split(",")[-1]) ]
    y_true = df_data['true output text']
    y_pred = df_data["generated text"]
    
    #ACC
    acc = sum(y_true==y_pred)/len(df_data)

    reference_sentences = [[eval(tt)] for tt in df_data['true output piece'].values.tolist()]
    candidate_sentences = [eval(tt) for tt in df_data['generated piece'].values.tolist()] 

    #BLEU n-gram
    bleu_score1, bleu_score2, bleu_score3, bleu_score4 = [], [], [], []
    for candidate, references in zip(candidate_sentences, reference_sentences):
        bleu_score1 = bleu_score1 + [sentence_bleu(references, candidate, weights=(1, 0, 0, 0), smoothing_function=smooth.method1)]
        bleu_score2 = bleu_score2 + [sentence_bleu(references, candidate, weights=(0, 1, 0, 0), smoothing_function=smooth.method1)] if len(candidate) >=2 else bleu_score2
        bleu_score3 = bleu_score3 + [sentence_bleu(references, candidate, weights=(0, 0, 1, 0), smoothing_function=smooth.method1)] if len(candidate) >=3 else bleu_score3
        bleu_score4 = bleu_score4 + [sentence_bleu(references, candidate, weights=(0, 0, 0, 1), smoothing_function=smooth.method1)] if len(candidate) >=4 else bleu_score4
    bleu_score1_avg = sum(bleu_score1)/len(bleu_score1)
    bleu_score2_avg = sum(bleu_score2)/len(bleu_score2)
    bleu_score3_avg = sum(bleu_score3)/len(bleu_score3)
    bleu_score4_avg = sum(bleu_score4)/len(bleu_score4)

    #Rouge
    hypothesis = [' '.join(tt ) for tt in candidate_sentences]
    reference = [' '.join(tt[0] ) for tt in reference_sentences]
    scores = rouge.get_scores(hypothesis, reference)

    rouge_1_scores = [score['rouge-1']['f'] for score in scores]
    rouge_2_scores = [score['rouge-2']['f'] for score, tt in zip(scores, reference) if len(tt.split()) >=2]
    rouge_l_scores = [score['rouge-l']['f'] for score in scores]
    rouge_1_avg = sum(rouge_1_scores) / len(rouge_1_scores)
    rouge_2_avg = sum(rouge_2_scores) / len(rouge_2_scores)
    rouge_l_avg = sum(rouge_l_scores) / len(rouge_l_scores)
    
    data = {'Task':task ,
            'dataset':len(df_data),
            'ACC':acc, 
            'BLEU-1':bleu_score1_avg, 'BLEU-2':bleu_score2_avg, 'BLEU-3':bleu_score3_avg, 'BLEU-4':bleu_score4_avg, 
            'Rouge_1':rouge_1_avg, 'Rouge_2':rouge_2_avg, 'Rouge_L':rouge_l_avg }
    df_result_gen = pd.concat([df_result_gen, pd.DataFrame(data, index=[0])])
    
df_result_gen = df_result_gen.round(4)
# df_result_gen.to_csv('out/metrics/result_table.csv', index=None)
print(df_result_gen)
```

# GC content

```python
df_GC = df[ df["input text"].apply(lambda x: "GC content" in x.split(",")[-1]) ]

import scipy.stats as stats
y_true = df_GC['true output text'].values.astype(float)
y_pred = df_GC["generated text"].apply(lambda x: x.split(',')[0]).values.astype(float)
plt.scatter(y_true,y_pred)
# 计算Pearson相关系数
pearson_corr, _ = stats.pearsonr(y_true, y_pred)
MAE = np.abs(y_true-y_pred).mean()
plt.title(f"GC content\nPearsonR={pearson_corr:.3f}, MAE={MAE:.3f}")
plt.gca().set_aspect('equal')
plt.xlabel("Ground-Truth GC Content")
plt.ylabel("Predicted GC Content")
# plt.savefig(f'out_20240102_balance/result/result_fig_regression_GCcontent.png', bbox_inches='tight', dpi=150)
```
